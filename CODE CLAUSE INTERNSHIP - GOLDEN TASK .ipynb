```python
import tkinter as tk
import speech_recognition as sr
import pyttsx3
import wikipedia
import webbrowser

# Initialize speech engine
engine = pyttsx3.init()

# Set up tkinter window
root = tk.Tk()
root.title("Voice Assistant")
root.geometry("500x300")

# Function to speak text
def speak(text):
    engine.say(text)
    engine.runAndWait()

# Function to recognize voice input
def record_and_process_voice():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        label_text.set("Listening...")
        root.update()  # Update the UI immediately
        audio = recognizer.listen(source)
        try:
            # Recognize the voice command
            command = recognizer.recognize_google(audio)
            label_text.set(f"Command: {command}")
            process_command(command)
        except sr.UnknownValueError:
            label_text.set("Sorry, I could not understand the command.")
        except sr.RequestError:
            label_text.set("Could not request results from Google Speech Recognition service.")
    
    root.update()  # Update the UI after processing

# Function to process the command
def process_command(command):
    command = command.lower()
    
    # Wikipedia search command
    if 'search' in command:
        search_term = command.replace('search', '').strip()
        label_text.set(f"Searching Wikipedia for: {search_term}")
        try:
            result = wikipedia.summary(search_term, sentences=2)
            label_text.set(result)
            speak(result)
        except wikipedia.exceptions.DisambiguationError as e:
            label_text.set(f"Multiple results found. Try to be more specific.")
            speak("Multiple results found. Try to be more specific.")
        except wikipedia.exceptions.PageError:
            label_text.set(f"Could not find anything about {search_term}")
            speak(f"Could not find anything about {search_term}")
    
    # Open a website
    elif 'open' in command:
        if 'google' in command:
            webbrowser.open('https://www.google.com')
            label_text.set("Opening Google")
            speak("Opening Google")
        elif 'youtube' in command:
            webbrowser.open('https://www.youtube.com')
            label_text.set("Opening YouTube")
            speak("Opening YouTube")
        else:
            label_text.set("I can only open Google and YouTube right now.")
            speak("I can only open Google and YouTube right now.")
    
    # Greet the user
    elif 'hello' in command or 'hi' in command:
        label_text.set("Hello! How can I assist you today?")
        speak("Hello! How can I assist you today?")
    
    # Exit the assistant
    elif 'exit' in command or 'bye' in command:
        label_text.set("Goodbye!")
        speak("Goodbye!")
        root.quit()
    
    # Unrecognized command
    else:
        label_text.set("Sorry, I don't understand that command.")
        speak("Sorry, I don't understand that command.")

# Create a simple UI using tkinter
label_text = tk.StringVar()
label_text.set("Click the microphone to speak")

# Label to show assistant responses
label = tk.Label(root, textvariable=label_text, wraplength=400, font=("Helvetica", 12))
label.pack(pady=20)

# Button to trigger voice command recording
mic_button = tk.Button(root, text="🎤 Microphone", command=record_and_process_voice, font=("Helvetica", 14))
mic_button.pack(pady=20)

# Run the tkinter main loop
root.mainloop()
```
